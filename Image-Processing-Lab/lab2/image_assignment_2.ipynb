```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
main_image = cv2.imread('/content/sample_data/sample_image.png',cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(cv2.cvtColor(main_image,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#mean kernel function

kernel_size = 7
n = (kernel_size -1)//2
image_size = 220
image = cv2.resize(main_image,(image_size,image_size))
result = np.zeros((image_size,image_size),dtype = 'float32')
for x in range(image_size):
  for y in range(image_size):
    sum = 0
    for i in range(kernel_size):
      for j in range(kernel_size):
        sum+= image[x-i-n,y-j-n]
    result[x,y] = sum / kernel_size **2
```

```python
plt.imshow(cv2.cvtColor(result,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
from skimage.exposure import rescale_intensity
out = rescale_intensity(result, in_range=(0, 255))

plt.imshow(cv2.cvtColor(out,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#meadian kernel function

kernel_size = 5
n = (kernel_size -1)//2
image_size = 420
image = cv2.resize(main_image,(image_size,image_size))
result = np.zeros((image_size,image_size),dtype = 'float32')
for x in range(image_size):
  for y in range(image_size):
    kr_value = []
    for i in range(kernel_size):
      for j in range(kernel_size):
        kr_value.append(image[x-i-n,y-j-n])
    kr_value.sort()
    mid = len(kr_value) //2
    result[x,y] = kr_value[mid]
```

```python
plt.imshow(cv2.cvtColor(result,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
from skimage.exposure import rescale_intensity
out = rescale_intensity(result, in_range=(0, 255))

plt.imshow(cv2.cvtColor(out,cv2.COLOR_BGR2RGB))
plt.show()
```
